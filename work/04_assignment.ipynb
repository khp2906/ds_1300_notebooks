```python
import matplotlib as plt
import pandas as pd
```

```python
df = pd.read_csv('../hw4/deaths15-49.csv')
df
```

```python
df = df[df['Country'].str.contains('South Korea')]
df
```

```python
#only want year, self-harm & neoplasms 
df1 = df[['Year', 'Self-harm', 'Neoplasms']]
df1
```

```python
# Plot
# Title, Legend, Axes or other Labels, and Color
```

```python
ax = df1.plot.scatter(x='Year', y='Self-harm', xlabel='Year', ylabel='Number of Deaths', c='#73c0ff', label = 'Self-harm', title='Cause of Deaths per Year')
df1.plot.scatter(ax = ax, x='Year', y='Neoplasms', xlabel='Year', ylabel='Number of Deaths', c='#ff7891', label = 'Neoplasms', title='Cause of Deaths per Year in South Korea (Ages 15-49)')
```
