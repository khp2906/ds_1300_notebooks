>Using the data below, answer the following questions:

>1) Which entities (top 5) had the largest population density in 2020?
>2) Which entities have more water area than land area?
>3) Which entities increased in population the most in the last 10 years?
>4) What state bird accounts for the largest population as of 2020? Land area?
>5) How many entities' largest city is their capital city?
>6) Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)

```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
#1: Which entities (top 5) had the largest population density in 2020?
facts['Population_density']=facts['Pop_2020']/facts['Area_land']
facts1 = facts.sort_values(by='Population_density',ascending=False)
facts1.head(5)
```

```python
#2: Which entities have more water area than land area?
facts2 = facts[facts['Area_land'] < facts['Area_water']]
facts2
```

```python
#3: Which entities increased in population the most in the last 10 years?
facts['Population_inc'] = facts['Pop_2020'] - facts['Pop_2010']
facts3 = facts.sort_values(by='Population_inc', ascending=False)
facts3.head(5)
```

```python
#4: What state bird accounts for the largest population as of 2020? Land area?
facts4bird = facts.sort_values(by='Pop_2020', ascending=False)
print(facts4bird['State_bird'].head(1))
print("")

facts4Larea = facts.sort_values(by='Area_land', ascending=False)
print(facts4Larea['State_bird'].head(1))
```

```python
#5: How many entities' largest city is their capital city?
count = 0
facts['Capital=LargestCity'] = facts['Capital'] == facts['City_1']
for i in range(56):
    if facts['Capital=LargestCity'][i] == True:
        count += 1
count
```

```python
#6: Which city has the largest percent drop from their largest city to their 5th largest? 
#   100*(1st largest - 5th largest)/(1st largest)
facts['LargestPercentDrop'] = (facts['city_1_pop'] - facts['city_5_pop']) / facts['city_1_pop'] * 100
facts6 = facts.sort_values(by='LargestPercentDrop', ascending=False)
facts6.head(1)
```

```python
#8: Which has the larger population density, the most dense US Territory or the least dense state?
mm['Status'].value_counts()

#find most dense US Territory
mmT = mm[mm['Status'] == 'Territory']
territory = mmT.sort_values(by='Population_density', ascending=False).head(1)

#find least dense State
mmS = mm[mm['Status'] == 'State']
state = mmS.sort_values(by='Population_density', ascending=True).head(1)

#compare territory and state
print("The most dense US Territory has a population density of ")
print(territory['Population_density'])
print("\nThe least dense US State has a population density of ")
print(state['Population_density'])
print("\nThe most dense US Territory, Puerto Ricohas, has the larger population density.")
```
