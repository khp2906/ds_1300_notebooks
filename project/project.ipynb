```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mcdonalds = pd.read_csv("../project/mcdDrinksMenu.csv") 
mcdonalds
```

```python
#edit out columns that we don't need
mcd = mcdonalds.drop(columns=['Calories from Fat', 'Total Fat (% Daily Value)', 'Saturated Fat', 
                              'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol', 'Cholesterol (% Daily Value)', 
                              'Sodium', 'Sodium (% Daily Value)', 'Carbohydrates (% Daily Value)', 'Dietary Fiber', 
                              'Dietary Fiber (% Daily Value)', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 
                              'Calcium (% Daily Value)', 'Iron (% Daily Value)'])
mcd
```

```python
#just want to look at the category coffeee & tea
mcd1 = mcd[mcd['Category'] == 'Coffee & Tea']
mcd1
```

```python
#only want size: 16 fl oz
mcd2 = mcd1[mcd1['Serving Size'] == '16 fl oz cup']
mcd2
```

```python
mcd2[mcd2['Item'].str.contains('|'.join(['Latte', 'Caramel', 'Coffee', 'Mocha']))]
```

```python
mcd3 = mcd2[~mcd2['Item'].str.contains('|'.join(['Iced', 'Tea', 'Hazelnut', 'Chocolate', 'Large', 'Vanilla']))]
mcd3
```

```python
mcd4 = mcd3[mcd3['Item'].str.contains('|'.join(['Caffè', 'Frappuccino']))]
mcd4
```

```python
mcd4.plot.bar(x='Item',  y=['Calories', 'Total Fat', 'Carbohydrates', 'Sugars'], colormap='plasma')
```

```python
starbucks = pd.read_csv("../project/starbucksDrinks.csv")
starbucks
```

```python
#to see column names
for col in starbucks.columns:
    print(col)
```

```python
#editing out columns we don't need
stb = starbucks.drop(columns=['Trans Fat (g) ', 'Saturated Fat (g)', 'Cholesterol (mg)', ' Dietary Fibre (g)', 
                              ' Protein (g) ', 'Vitamin A (% DV) ', 'Vitamin C (% DV)', ' Calcium (% DV) ', 
                              'Iron (% DV) ', 'Caffeine (mg)'])
stb
```

```python
stb['Beverage_category'].value_counts()
```

```python
#don't need smoothies, iced drinks, tea
stb1 = stb[~stb['Beverage_category'].str.contains('|'.join(['Smoothies', 'Shaken Iced', 'Tea', 'Light', 'Crème']))]
stb1
```

```python
#only need nonfat milk
stb2 = stb1[~stb1['Beverage_prep'].str.contains('|'.join(['Whole', 'Soymilk', 'Whole', '2%']))]
stb2
```

```python
#only want size: grande
stb3 = stb2[~stb2['Beverage_prep'].str.contains('|'.join(['Tall', 'Venti', 'Short']))]
stb3
```

```python
#taking out drinks we don't need
stb4 = stb3[~stb3['Beverage'].str.contains('|'.join(['Vanilla', 'Brewed', 'Cappuccino', 'Espresso', 'Skinny', 'Macchiato',
                                                     'Chocolate', 'Apple', 'Coffee', 'Java']))]
stb4
```

```python
#picking drinks we need
stb5 = stb4[stb4['Beverage'].str.contains('|'.join(['Caffè', 'Caramel']))]
stb5
```

```python
stb5.plot.bar(x='Beverage',  y=['Calories', ' Total Fat (g)', ' Total Carbohydrates (g) ', ' Sugars (g)'], colormap='plasma')
```

```python
mcd5 = mcd4.drop(columns=['Category', 'Protein'])
mcd6 = mcd5.rename(columns={"Calories": "Calories (g)", "Total Fat": "Fat (g)", "Carbohydrates": "Carbohydrates (g)", 
                            "Sugars": "Sugars (g)"})
mcd6 
mcd6['Fat (g)'] = mcd6['Fat (g)'].astype(float)
mcd6
```

```python
stb5
stb6 = stb5.reindex([42, 10, 22, 202])
stb7 = stb6.drop(columns=['Beverage_category', ' Sodium (mg)'])
stb8 = stb7.rename(columns={"Beverage": "Item", "Beverage_prep": "Serving Size", "Calories": "Calories (g)", 
                            " Total Fat (g)": "Fat (g)", " Total Carbohydrates (g) ": "Carbohydrates (g)", 
                            " Sugars (g)": "Sugars (g)"})
stb8['Fat (g)'] = stb8['Fat (g)'].astype(float)
stb8
```

```python
sns.set_theme(style="white")
stb8['Place']='Starbucks'
mcd6['Place']="McDonald's"
res=pd.concat([stb8,mcd6])
sns.barplot(x='Item',y='Fat (g)', data=res,hue="Place")
```
